```python
import pandas as pd
fileURL = "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(fileURL, names=["Sepal Length", "Sepal Width",
                                 "Petal Length", "Petal Width",
                                 "Name"], header=None)
```

```python
df.head()
```

```python
print(" Question 1.1- Summary Statistics\n There are 4 features. The data types for each feature is as follows: \n")
print(" Sepal Length: numerical \n Sepal Width: numerical\n Petal Length: numerical \n Pedal Width: numerical")
print(" Name: class")
```

```python
summary = pd.DataFrame({"Research Category":[ "Sepal Length", "Sepal Width", "Petal Length", "Petal Width"], 
                   "count":[df['Sepal Length'].count(),df['Sepal Width'].count(),df['Petal Length'].count(),df['Petal Width'].count()],
                   "mean":[df['Sepal Length'].mean(),df['Sepal Width'].mean(),df['Petal Length'].mean(),df['Petal Width'].mean()],
                   "standard_deviation":[df['Sepal Length'].std(),df['Sepal Width'].std(),df['Petal Length'].std(),df['Petal Width'].std()],     
                   "min":[df['Sepal Length'].min(), df['Sepal Width'].min(), df['Petal Length'].min(), df['Petal Width'].min()], 
                   "25 Percentile":[df['Sepal Length'].quantile(.25),df['Sepal Width'].quantile(.25),df['Petal Length'].quantile(.25),df['Petal Width'].quantile(.25)],
                   "50 Percentile":[df['Sepal Length'].quantile(.50),df['Sepal Width'].quantile(.50),df['Petal Length'].quantile(.50),df['Petal Width'].quantile(.50)],
                   "75 Percentile":[df['Sepal Length'].quantile(.75),df['Sepal Width'].quantile(.75),df['Petal Length'].quantile(.75),df['Petal Width'].quantile(.75)],
                   "max":[df['Sepal Length'].max(), df['Sepal Width'].max(), df['Petal Length'].max(),df['Petal Width'].max()],
                   "range":[df['Sepal Length'].max() - df['Sepal Length'].min(), df['Sepal Width'].max() - df['Sepal Width'].min(), df['Petal Length'].max() - df['Petal Length'].min(), df['Petal Width'].max() - df['Petal Width'].min()],
                   "variance":[df['Sepal Length'].var(),df['Sepal Width'].var(),df['Petal Length'].var(),df['Petal Width'].var()],
                    })


summary.set_index('Research Category', inplace=True)


```

```python
print("Below is a summary statistics of the Iris Dataset")
summary.transpose()
```
