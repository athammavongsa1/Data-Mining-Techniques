```python
import pandas as pd
import numpy as np 
import sklearn
from sklearn.metrics import mean_squared_error

from sklearn.model_selection import train_test_split

fileURL = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(fileURL, names =['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width','Species']
                 , header = None )
iris = iris.dropna()


```

```python
%pylab inline

# (c) 2014 Reid Johnson and Everaldo Aguiar
#
# Functions to work with continuous data and linear regression models.

import matplotlib.pyplot as pl

def pairs(data):
    """Generates and shows a pairwise scatterplot of the dataset features.

    A figure with nxn scatterplots is generated, where n is the number of features. The features are
    defined as the all columns excluding the final column, which is defined as the class.

    Args:
      data (array): A dataset.

    """
    i = 1

    # Divide columns into features and class.
    features = list(data.columns)
    classes = features[-1] # create class column
    del features[-1] # delete class column from feature vector

    # Generate an nxn subplot figure, where n is the number of features.
    figure = pl.figure(figsize=(5*(len(data.columns)-1), 4*(len(data.columns)-1)))
    for col1 in data[features]:
        for col2 in data[features]:
            ax = pl.subplot(len(data.columns)-1, len(data.columns)-1, i)
            if col1 == col2:
                ax.text(2.5, 4.5, col1, style='normal', fontsize=20)
                ax.axis([0, 10, 0, 10])
                pl.xticks([]), pl.yticks([])
            else:
                for name in data[classes]:
                    cond = data[classes] == name
                    ax.plot(data[col2][cond], data[col1][cond], linestyle='none', marker='o', label=name)
                #t = plt.title(name)
            i += 1

    pl.show()
```

A scatterplot matrix of scatterplots, with one scatterplot for each pairwise combination
of features.

```python
pairs(iris)
```

Going to divide the data into training and testing sets

```python
train, test = sklearn.model_selection.train_test_split(iris, test_size=.2)
```

You're going to run linear regression on train and view it against test. If, when the algo hasn't seen the test data, it lets you know how well the algorithm generalizes later on in unseen data. That's why it's split between train and test. 

A plot of the linear regression models generated on each pairwise combination of features,
with corresponding model statistics

```python
#"This tests linear regression between x = sepal length and y = sepal width" 
#"This establishes the x and y train and tests."
from sklearn.linear_model import LinearRegression
x= train['Sepal Length']
y= train['Sepal Width']
x_test = test['Sepal Length']
y_test = test['Sepal Width']


#"This reshapes the values to work with .fit()"

x=x.values.reshape(-1,1)
y=y.values.reshape(-1,1)
y_test = y_test.values.reshape(-1,1)
x_test = x_test.values.reshape(-1,1)

#"This fits the graph returning object reg. "
reg = LinearRegression().fit(x, y)
y_predict = reg.predict(x_test)



#"This gives us metrics"
# The coefficients
# The coefficient of determination: 1 is perfect prediction
print('Coefficients:', reg.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_predict))
# Explained variance score: 1 is perfect prediction
print("Variance score : %.2f"% reg.score(x_test , y_test))


#"This plots the graph"
plt.scatter(x_test, y_test,  color='black')
plt.plot(x_test, y_predict, color='purple', linewidth=3)

    
plt.xticks(())
plt.yticks(())
plt.suptitle("Sepal Length and Sepal Width", size=16)
plt.show()


```

Generally, overfitting occurs when the model models the training set too well-- and interprets noise and fluctuations of the dataset as a concept  of the model and tends to happen when you have too many features. If you remove a feature, you may begin to underfit the model or the model cannot generate new data or model its training data because its too simple and does not learn from the training data.
